```python
import pandas as pd
from jinja2 import Template
```

```python
df = pd.read_excel("output_excel/S01_data.xlsx")
```

```python
df.head()
```

```python
# role = "Chair"
role = "Co-chair"

filter_role = df["Role"] == role
df_role = df[filter_role]
```

```python
for data in df_role.to_dict(orient="records")[:]:
    
    # Read HTML template
    with open(data["template_html"], "r", encoding="utf-8") as f:
        html = f.read()
    
    # Render template
    template = Template(html)
    rendered = template.render(**data)

    # Save or use rendered HTML
    with open(f"{data['html_path']}/{data['html']}", "w", encoding="utf-8") as f:
        f.write(rendered)
        
    print(f"Generated {data['html']} for {data['Name']}")
```
