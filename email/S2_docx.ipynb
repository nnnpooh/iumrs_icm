```python
import pandas as pd
from docxtpl import DocxTemplate
```

```python
df = pd.read_excel("output_excel/S01_data.xlsx")
```

```python
df.head()
```

```python
for row in df.iloc[:1].to_dict(orient="records"):
    print(row)
```

```python
filt = df["Role"] == "Chair"

df_chair = df[filt]
template_path = "templates/Invitation IUMRS-ICEM2026_Chair_v2_template.docx"
for context in df_chair.to_dict(orient="records")[:]:
    tpl = DocxTemplate(template_path)
    tpl.render(context)
    tpl.save(f'output_docx_chair/{context["docx"]}')
```
