```python
import pandas as pd
from docxtpl import DocxTemplate
```

```python
df = pd.read_excel("output_excel/S01_data.xlsx")
```

```python
df.head()
```

```python
for row in df.iloc[:1].to_dict(orient="records"):
    print(row)
```

```python
# role = "Chair"
role = "Co-chair"

filt = df["Role"] == role
df_role = df[filt]
print(df_role["Role"].value_counts())
```

```python
for data in df_role.to_dict(orient="records")[:]:
    tpl = DocxTemplate(data["template_docx"])
    tpl.render(data)
    tpl.save(f'{data["docx_path"]}/{data["docx"]}')
    print(f"Saved {data['docx']} to {data['docx_path']}")
```
