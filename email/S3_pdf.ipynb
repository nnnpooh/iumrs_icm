```python
import pandas as pd
import os
from docx2pdf import convert
```

```python
# role = "Chair"
role = "Co-chair"
```

```python
if role == "Chair":
    for file in os.listdir("output_docx/chair"):
        if file.endswith(".docx"):
            convert(
                os.path.join("output_docx/chair", file),
                os.path.join("output_pdf/chair", file.replace(".docx", ".pdf")),
            )

elif role == "Co-chair":
    for file in os.listdir("output_docx/cochair"):
        if file.endswith(".docx"):
            convert(
                os.path.join("output_docx/cochair", file),
                os.path.join("output_pdf/cochair", file.replace(".docx", ".pdf")),
            )
```
