```python
import pandas as pd
import os
from docx2pdf import convert
```

```python
# role = "Chair"
role = "Co-chair"
```

```python
if role == "Chair":
    for file in os.listdir("output_docx/chair"):
        if file.endswith(".docx"):
            convert(
                os.path.join("output_docx/chair", file),
                os.path.join("output_pdf/chair", file.replace(".docx", ".pdf")),
            )

elif role == "Co-chair":
    for file in os.listdir("output_docx/cochair"):
        if file.endswith(".docx"):
            convert(
                os.path.join("output_docx/cochair", file),
                os.path.join("output_pdf/cochair", file.replace(".docx", ".pdf")),
            )
```

```python
from PyPDF2 import PdfReader

# role = "Chair"
role = "Co-chair"

def get_pdf_page_count(path):
    reader = PdfReader(path)
    return len(reader.pages)

if role == "Chair":
    for file in os.listdir("output_pdf/chair"):
        if file.endswith(".pdf"):
            print(file, get_pdf_page_count(os.path.join("output_pdf/chair", file)))
elif role == "Co-chair":
    for file in os.listdir("output_pdf/cochair"):
        if file.endswith(".pdf"):
            print(file, get_pdf_page_count(os.path.join("output_pdf/cochair", file)))
```
