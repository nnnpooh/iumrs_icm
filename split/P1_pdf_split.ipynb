```python
import pandas as pd
import os
from pypdf import PdfReader, PdfWriter
```

```python
df = pd.read_excel('data.csv')
```

```python
def split_pdf_to_single_pages(input_pdf_path, output_folder):
    """Splits a PDF into individual single-page PDF files."""
    # Ensure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the original PDF file in read binary mode
    with open(input_pdf_path, 'rb') as infile:
        reader = PdfReader(infile)
        num_pages = len(reader.pages)
        base_filename = os.path.splitext(os.path.basename(input_pdf_path))[0]

        for page_num in range(num_pages):
            # Create a new PdfWriter object for each page
            writer = PdfWriter()
            # Add the current page to the writer
            writer.add_page(reader.pages[page_num])

            # Define the output filename
            output_filename = f"{output_folder}/{base_filename}_page_{page_num + 1}.pdf"

            # Save the new single-page PDF file
            with open(output_filename, 'wb') as outfile:
                writer.write(outfile)
            print(f"Created: {output_filename}")

# Usage example:
# split_pdf_to_single_pages("document.pdf", "output_pages")
```
